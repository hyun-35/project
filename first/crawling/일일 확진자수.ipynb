## 확진자수 크롤링

```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome('C:/Users/L/수업실습/chromedriver/chromedriver.exe')
url = 'http://ncov.mohw.go.kr/'
driver.get(url)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

### 1. 지역명 가져오기

```python
numbers = soup.select('div.rpsam_graph > div#main_maplayout2')[0]
```

```python
num_data1 = []

v = numbers.select('span.name')

for i in v:
    local_num = i.text
    num_data1.append([local_num])
```

```python
import pandas as pd
columns = ['지역명']
pd_data1 = pd.DataFrame(num_data1, columns = columns)
pd_data1
```

### 2. 확진자수 가져오기

```python
num_data2 = []

v = numbers.select('span.before')

for i in v:
    local_num = i.text
    num_data2.append([local_num])
```

```python
num_data2
```

```python
num_data3 = []

for i in num_data2:
    for j in i:
        j1 = j.strip("'(+")
        j2 = j1.strip(")'")
        j3 = int(j2)
        num_data3.append(j3)
```

```python
num_data3
```

```python
import pandas as pd
columns = ['확진자수']
pd_data2 = pd.DataFrame(num_data3, columns = columns)
pd_data2
```

```python
today_number = pd_data1.join(pd_data2)
#today_number1 = today_number.sort_values('확진자수', ascending = False)
#today_number1
```

```python
d = today_number[today_number['지역명'] =='검역'].index
today_number2 = today_number.drop(d)
today_number2
```

```python
today_number2 = today_number2.sort_values('확진자수', ascending = False)
today_number2
```

### 3. 확진자수 순위 매기기

```python
num = list(range(1,18))
today_number2['순위'] = num
today_number2
```

```python
today_number2 = today_number2.sort_values('지역명')
```

### 4. 날짜(today) 추가

```python
from datetime import datetime

today = datetime.today()
month = today.month
day = today.day
date = str(month) + '/' + str(day)
date
```

```python
day_list = []
a = 17
while a:
    day_list.append(date)
    a = a-1
```

```python
pd_data5 = pd.DataFrame(day_list, columns = ['날짜'])
```

```python
today_number3 = today_number2.join(pd_data5)
```

```python
today_number3.to_csv('./확진자수/today_number0913.csv', index = False)
```
