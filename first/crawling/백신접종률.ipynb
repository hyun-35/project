## 백신접종률 크롤링

```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome('C:/Users/L/수업실습/chromedriver/chromedriver.exe')
url = 'https://search.daum.net/search?w=tot&DA=YZR&t__nil_searchbox=btn&sug=&sugo=&sq=&o=&q=%EB%B0%B1%EC%8B%A0+%EC%A0%91%EC%A2%85+%ED%98%84%ED%99%A9'
driver.get(url)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

### 1. 지역명 가져오기

```python
v_percent = soup.select('div:nth-child(4) > div.cont_info > div._cont_map > div:nth-child(3)')[0]
```

```python
per_data1 = []

v = v_percent.select('li > a > span.txt_location')

for i in v:
    local_name = i.text
    per_data1.append([local_name])
```

```python
import pandas as pd
columns = ['지역명']
pd_data1 = pd.DataFrame(per_data1, columns = columns)
pd_data1
```

### 2. 백신접종률 가져오기

```python
per_data2 = []

pers = v_percent.select('li > a > strong.txt_rate')

for per in pers:
    percent = per.text
    percent = percent.strip('%')
    per_data2.append([percent])
```

```python
import pandas as pd
columns = ['백신접종률']
pd_data2 = pd.DataFrame(per_data2, columns = columns)
pd_data2
```

```python
local_per = pd_data1.join(pd_data2)
local_per
```

```python
local_per2 = local_per.sort_values('지역명')
local_per2
```

### 3. 날짜(today) 추가

```python
from datetime import datetime

today = datetime.today()
month = today.month
day = today.day
date = str(month) + '/' + str(day)
date
```

```python
day_list = []
a = 17
while a:
    day_list.append(date)
    a = a-1
```

```python
pd_data5 = pd.DataFrame(day_list, columns = ['날짜'])
```

```python
local_per3 = local_per2.join(pd_data5)
```
