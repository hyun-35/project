## 거리두기 단계 크롤링

```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome('C:/Users/L/수업실습/chromedriver/chromedriver.exe')
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EA%B1%B0%EB%A6%AC%EB%91%90%EA%B8%B0+%EB%8B%A8%EA%B3%84'
driver.get(url)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

### 1. 지역명 가져오기

```python
local_data1 = []

names = soup.select('div.local_info > strong.local_name')

for name in names:
    local_name = name.text
    local_data1.append([local_name])
```

```python
import pandas as pd
columns = ['지역명']
pd_data1 = pd.DataFrame(local_data1, columns = columns)
```

```python
data1 = pd_data1[1:]
data1
```

### 2. 거리두기 단계

```python
local_data2 = []

numbers = soup.select('div.local_info > p.population_number')

for number in numbers:
    local_number = number.text
    local_data2.append([local_number])
    
columns = ['거리두기 단계']
pd_data2 = pd.DataFrame(local_data2, columns = columns)   
```

```python
data2 = pd_data2[1:]
data2
```

```python
local_number = data1.join(data2)
```

```python
local_number
```

```python
local_number2 = local_number.drop_duplicates(subset = '지역명')
local_number2
```

```python
local_number3 = local_number2.sort_values('지역명')
local_number3
```

```python
local_number4 = local_number3.reset_index(drop=True)
local_number4
```

### 3. 날짜(today) 컬럼 추가

```python
from datetime import datetime
```

```python
today = datetime.today()
month = today.month
day = today.day
date = str(month) + '/' + str(day)
date
```

```python
day_list = []
a = 17
while a:
    day_list.append(date)
    a = a-1
```

```python
pd_data5 = pd.DataFrame(day_list, columns = ['날짜'])
```

```python
pd_data5
```

```python
local_number5 = local_number4.join(pd_data5)
local_number5
```

```python
local_number5.to_csv('./거리두기단계/local_number0908.csv', index = False)
```
